## ¿Qué es la Programación Orientada a Objetos?

La Programación Orientada a Objetos (POO) es un paradigma de programación que organiza el código en torno a "objetos". Un objeto es una instancia de una "clase", que define las propiedades (atributos) y comportamientos (métodos) que el objeto puede tener.

### Ventajas de POO
- **Modularidad**: Las clases permiten organizar el código en unidades independientes.
- **Abstracción**: Facilita modelar entidades del mundo real.
- **Encapsulamiento**: Protege los datos al ocultar detalles internos.
- **Herencia**: Permite reutilizar código mediante la creación de nuevas clases basadas en existentes.
- **Polimorfismo**: Habilita el uso uniforme de objetos de diferentes clases.

### Conceptos Básicos
- **Clase**: Plantilla que define atributos y métodos.
- **Objeto**: Instancia de una clase con valores específicos para sus atributos.

## Clases y Objetos

En Python, las clases se definen con la palabra clave `class`. Los objetos se crean instanciando una clase.

### Definición de una Clase
Una clase se define con `class` seguido del nombre y dos puntos. El cuerpo contiene atributos y métodos.

### Creación de Objetos
Un objeto se crea llamando al nombre de la clase como si fuera una función.

````
class Perro:
    # Atributo de clase
    especie = "Canino"

    # Definimos el método constructor
    def __init__(self, nombre, edad):
        # Atributos de instancia
        self.nombre = nombre
        self.edad = edad

    # Método
    def ladrar(self):
        print(f"{self.nombre} está ladrando.")


# Creación de objetos
perro1 = Perro("Fido", 3)
perro2 = Perro("Rex", 5)


# Acceso a atributos y métodos
print(perro1.nombre)  # Salida: Fido
print(perro2.edad)    # Salida: 5
perro1.ladrar()       # Salida: Fido está ladrando.
````

````
print(perro1)
````

````
class Coche:

  def __init__(self, marca, modelo):
      self.marca = marca
      self.modelo = modelo

  def conducir(self):
      print(f"El coche {self.marca} {self.modelo} está siendo conducido.")

  # si no tiene __str__ entonces usará: __repr__
  # Sin print en el interprete interactivo usa __repr__
  def __str__(self):
      return f"{self.marca} {self.modelo}"

  def __repr__(self):
      return f"Coche('{self.marca}', '{self.modelo}')"


# Crear un objeto y probar el método
mi_coche = Coche("Toyota", "Corolla")
print(mi_coche)

mi_coche.conducir()  # Salida: El coche Toyota Corolla está siendo conducido.
````

### Ejercicio: Calculadora con métodos y atributos
Crea una clase llamada Calculadora que permita realizar las operaciones básicas: suma, resta, multiplicación y división.

### Requisitos:

> La clase debe tener dos atributos: numero1 y numero2.

> Debe incluir métodos para:

> Sumar los dos números.

> Restar los dos números.

> Multiplicar los dos números.

> Dividir los dos números (si numero2 es 0, mostrar un mensaje de error).

> Crear una instancia de la clase y demostrar el uso de cada método.



````
class Calculadora:
    def __init__(self, numero1, numero2):
        self.numero1 = numero1
        self.numero2 = numero2

    def sumar(self):
        return self.numero1 + self.numero2

    def restar(self):
        return self.numero1 - self.numero2

    def multiplicar(self):
        return self.numero1 * self.numero2

    def dividir(self):
        if self.numero2 == 0:
            return "Error: No se puede dividir por cero."
        return self.numero1 / self.numero2

# Uso de la clase
calc = Calculadora(10, 0)

print("Suma:", calc.sumar())
print("Resta:", calc.restar())
print("Multiplicación:", calc.multiplicar())
print("División:", calc.dividir())
````

## Herencia

La herencia permite crear nuevas clases basadas en clases existentes.

### Clases Base y Derivadas
- **Clase Base**: Clase de la que se hereda.
- **Clase Derivada**: Clase que hereda.

### Sobrescritura de Métodos
Las clases derivadas pueden redefinir métodos de la clase base.

A continuación, se muestra un diagrama simple de herencia:

```
Vehiculo
  ├── Coche
  └── Bicicleta
```

````
# Clase Base (Padre)
class Vehiculo:
    def __init__(self, marca, modelo, ruedas):
        self.marca = marca
        self.modelo = modelo
        self.ruedas = ruedas

    def info(self):
        return f"Marca: {self.marca}, Modelo: {self.modelo}, Ruedas: {self.ruedas}"

    def arrancar(self):
        return f"{self.modelo} está arrancando..."

    def frenar(self):
        return f"{self.modelo} está frenando...     "



# Clase Hija 1: Coche
class Coche(Vehiculo):
    def __init__(self, marca, modelo, puertas, ruedas=4):
        super().__init__(marca, modelo, ruedas)  # Llamamos al constructor de la clase base
        self.puertas = puertas  # Propiedad propia de los coches

    def info(self):
        # Podemos sobrescribir el método de la clase base para añadir más detalles
        return f"{super().info()}, Puertas: {self.puertas}"

    def encender_aire_acondicionado(self):
        return f"El aire acondicionado del {self.modelo} está encendido."



# Clase Hija 2: Bicicleta
class Bicicleta(Vehiculo):
    def __init__(self, marca, modelo, tipo, ruedas=2):
        super().__init__(marca, modelo, ruedas)
        self.tipo = tipo  # Propiedad propia de las bicicletas

    def info(self):
        # Sobrescribimos el método para que también incluya el tipo de bicicleta
        return f"{super().info()}, Tipo: {self.tipo}"

    def hacer_wheelie(self):
        return f"¡El {self.modelo} está haciendo wheelie!"
````

````
# Instanciamos la clase Coche (subclase de Vehiculo)
coche = Coche("Toyota", "Corolla", puertas=4)

print(coche.marca)                          # atributo
print(coche.arrancar())                     # método heredado de Vehiculo
print(coche.encender_aire_acondicionado())  # método propio
````

````
# Instanciamos la clase Bicicleta (subclase de Vehiculo)
bicicleta = Bicicleta("Trek", "Marlin", "Montaña")

print(bicicleta.tipo)                  # atributo
print(bicicleta.frenar())              # método heredado
print(bicicleta.hacer_wheelie())       # método propio
````

### Ejercicio Práctico
Crea una clase base `Vehiculo` con un método `mover`, y dos clases derivadas `Coche` y `Bicicleta` que sobrescriban `mover`.


````
# @title Solución
class Vehiculo:
    def mover(self):
        return f"El vehículo se está moviendo."

class Coche(Vehiculo):
    def mover(self):
        return f"El coche está conduciendo."

class Bicicleta(Vehiculo):
    def mover(self):
        return f"La bicicleta está pedaleando."

# Uso
coche = Coche()
bicicleta = Bicicleta()

print(coche.mover())      # Salida: El coche está conduciendo.
print(bicicleta.mover())  # Salida: La bicicleta está pedaleando.
````

## Proyecto Final: Sistema de Gestión de Biblioteca

Este proyecto implementa un sistema básico de gestión de biblioteca utilizando POO.

### Descripción
El sistema incluye clases para `Libro`, `Usuario` y `Biblioteca`, con funcionalidades para agregar libros, prestarlos y devolverlos.

### Implementación

````
class Libro:
    def __init__(self, titulo, autor, isbn):
        self.titulo = titulo
        self.autor = autor
        self.isbn = isbn
        self.disponible = True

    def __str__(self):
        return f"{self.titulo}, {self.autor}"

    def __repr__(self):
        return f"Libro({self.titulo}, {self.autor})"

    def prestar(self):
        if self.disponible:
            self.disponible = False
            return True
        return False

    def devolver(self):
        self.disponible = True



class Usuario:
    def __init__(self, nombre, id_usuario):
        self.nombre = nombre
        self.id_usuario = id_usuario
        self.libros_prestados = []

    def __str__(self):
        return f"{self.nombre}, {self.id_usuario}"

    def __repr__(self):
        return f"Usuario({self.nombre}, {self.id_usuario})"

    def tomar_prestado(self, libro):
        if libro.prestar():
            self.libros_prestados.append(libro)
            return True
        return False

    def devolver_libro(self, libro):
        if libro in self.libros_prestados:
            libro.devolver()
            self.libros_prestados.remove(libro)
            return True
        return False



class Biblioteca:
    def __init__(self, nombre):
        self.nombre = nombre
        self.libros = {}
        self.usuarios = {}

    def __repr__(self):
        return f"Biblioteca({self.nombre})"

    def agregar_libro(self, libro):
        # Guardamoss como valor el objeto libro
        self.libros[libro.isbn] = libro

    def registrar_usuario(self, usuario):
        self.usuarios[usuario.id_usuario] = usuario

    def prestar_libro(self, isbn, id_usuario):
        # El método get de diccionarios devuelve el valor
        libro = self.libros.get(isbn)
        usuario = self.usuarios.get(id_usuario)

        if libro and usuario:
            return usuario.tomar_prestado(libro)

        return False

    def devolver_libro(self, isbn, id_usuario):
        usuario = self.usuarios.get(id_usuario)

        if usuario:
            for libro in usuario.libros_prestados:
                if libro.isbn == isbn:
                    return usuario.devolver_libro(libro)

        return False
````

### Probando nuestra Biblioteca

````
# Creamos la biblioteca
biblioteca = Biblioteca("Biblioteca Central")
````

````
# Creamos objetos tipo Libro
libro1 = Libro("El Quijote", "Cervantes", "12345")
libro2 = Libro("Cien Años de Soledad", "García Márquez", "67890")
````

````
# Agregamos libros a la biblioteca
biblioteca.agregar_libro(libro1)
biblioteca.agregar_libro(libro2)
````

````
# Consultamos si se agregaron los libros
print(biblioteca.libros)
````

````
# Creamos un usuario y lo registramos a la biblioteca
usuario1 = Usuario("Juan", "U001")
biblioteca.registrar_usuario(usuario1)
````

````
# Mostramos los usuarios de la biblioteca
print(biblioteca.usuarios)
````

````
# La Biblioteca le presta un libro al usuario
biblioteca.prestar_libro("12345", "U001")
print(usuario1.libros_prestados[0].titulo)  # Salida: El Quijote
````

````
# Consultamos cuantos libros tiene el usuario1
print(f"Libros que tiene el usuario: {len(usuario1.libros_prestados)}")

# La Biblioteca recibe el libro del usuario
biblioteca.devolver_libro("12345", "U001")
print("\n¡Se devolvió un libro!")

# Consultamos cuantos libros tiene ahora el usuario1
print(f"\nLibros que tiene el usuario: {len(usuario1.libros_prestados)}")
````
