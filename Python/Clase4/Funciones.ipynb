## Ejercicio — Escalado entre 0 y 1

El **escalado Min-Max** transforma los datos para que queden dentro del rango:

$$
0 \le x' \le 1
$$

La fórmula es:

$$
x' = \frac{x - \min(x)}{\max(x) - \min(x)}
$$

donde:

- $\min(x)$ = valor mínimo de los datos  
- $\max(x)$ = valor máximo de los datos  

---

datos = [59, 71, 62, 23, 49, 58, 64]

*   Encuentre el valor mínimo y el valor máximo.
*   Aplique la fórmula de escalado a cada valor.
*   Devuelva una nueva lista con los valores escalados.

````python
# @title
def escalar(datos):
    escalados = []
    minimo = min(datos)
    maximo = max(datos)

    return [(x - minimo) / (maximo - minimo) for x in datos]


# Por afuera de la función creamos la lista
datos = [59, 71, 62, 23, 49, 58, 64]

# imprimimos lo que devuelve la función
print(escalar(datos))
````

## Concepto de Pase de argumentos en Python

En Python, los argumentos se pasan como **referencias a objetos** (modelo *pass-by-object-reference*).  
Esto significa que la función recibe una referencia al mismo objeto, no una copia.


*   `Objetos mutables`

Si el objeto se **modifica dentro de la función**, el cambio se refleja fuera,  
porque ambas variables apuntan al **mismo objeto**.


*   `Objetos inmutables`

No se pueden modificar. Cualquier “cambio” dentro de la función en realidad  
**crea un nuevo objeto**, por lo que el valor original **no cambia** fuera de la función.

<img src='https://files.realpython.com/media/Mutable-and-Immutable-Data-Types-in-Python_Watermarked.2566c67a719d.jpg' width=600 />

*Los inmutables no cambian dentro de la función; se crea un nuevo objeto.*

````python
# Objetos inmutables (int)
def agregar_elemento(x):
    x = x + 4

# Por afuera de la función:
numero = 3
agregar_elemento(numero)
print("Número después de la función:", numero)
````

*Los mutables sí pueden cambiar dentro de la función.*

````python
# Objetos mutables (lista)
def agregar_elemento(lst):
    lst.append(4)


mi_lista = [1, 2, 3]
agregar_elemento(mi_lista)
print("Lista después de la función:", mi_lista)
````

## Recursividad
La recursividad es una técnica de programación en la que una función se llama a sí misma para resolver un problema. Esta técnica se utiliza para dividir un problema en subproblemas más pequeños, y se basa en dos principios clave:<br><br>

* Caso Base: La condición que detiene la recursión. Sin un caso base, la función
seguirá llamándose a sí misma indefinidamente.<br><br>
* Caso Recursivo: La parte de la función que se llama a sí misma para trabajar en una versión más pequeña del problema.

````python
# Función para calcular el factorial de n
def factorial(n):
    # Caso base
    if n <= 1:
        return 1
    # Caso recursivo
    else:
        return n * factorial(n - 1)

# Ejemplo de uso
print(factorial(5))
````

## Funciones generadoras

Las funciones generadoras son un tipo especial de función en Python que permiten generar una secuencia de valores a lo largo del tiempo, en lugar de devolverlos todos de una vez. Esto es especialmente útil cuando se trabaja con grandes conjuntos de datos o cuando solo se necesita un valor a la vez.

Las funciones generadoras se definen como las funciones regulares pero usan la palabra clave yield en lugar de return. Cada vez que la función generadora encuentra un yield, guarda su estado actual y devuelve el valor. La próxima vez que se invoque la función, continuará desde donde se detuvo.

````python
# Generador de números pares
def generador_pares(limite):
    num = 0
    while num <= limite:
        yield num

        num += 2
````

````python
# De qué tipo es?
print(type(generador_pares(10)))
````

````python
# Si queremos correrlo de esta manera, no obtenemos nada
generador_pares(10)
````

*`generador_pares(10) `crea el generador, pero no genera todos los valores todavía*

````python
# generador_pares() crea el generador
pares = generador_pares(10)

# Ahora debemos consumirlo con next
print(next(pares))
print(next(pares))
print(next(pares))
print(next(pares))
print(next(pares))
print(next(pares))
````

````python
pares = generador_pares(10)

for _ in range(7):
  print(next(pares))
````

## Expresiones generadoras (Generator Expressions)

Las **expresiones generadoras** crean un **generador** de forma compacta, parecida a una lista por comprensión, pero usando **paréntesis** en lugar de corchetes.

### Características clave
- Se evalúan **bajo demanda** (lazy evaluation)
- Usan **menos memoria** que las listas
- No almacenan todos los valores, los producen uno por uno
- No necesitan `yield` (eso es solo para funciones generadoras)

### Sintaxis
```python
(expresion for elemento in iterable if condicion)


````python
expr = (i for i in range(12) if i%2==0)

print(next(expr))
print(next(expr))
print(next(expr))
print(next(expr))
print(next(expr))
print(next(expr))
````

````python
# También podemos usar
mi_expr = sum(i for i in range(10) if i%2==0) # <=== acá sum le pide al generador el siguiente valor.
print(mi_expr)
````

````python
# Expresión generadora
cuadrados = (x**2 for x in range(5))

# Podemos usar next si queremos obtener un solo elemento:
primer_valor = next(cuadrados) # 0
print(primer_valor)

# Si queremos el resto, el for se encarga del next():
for n in cuadrados:
    print(n) # 1, 4, 9, 16
````

## Comprensión de Listas en Python
### *(List Comprehension)*
La comprensión de listas es una forma concisa y poderosa de crear listas en Python. Permite construir listas de manera rápida a partir de iterables existentes (como listas, tuplas, o cualquier objeto iterable) aplicando una expresión y, opcionalmente, condiciones.

````python
'''
variable = [nueva_expresión for elemento in iterable if condición]
'''
````

````python
# Ejemplo para crear pares e impares con list comprehension
pares = [i for i in range(100) if i%2 == 0]
impares = [i for i in range(100) if i%2 == 1]

print(pares)
print(impares)
````

# Ejercicio: Estandarización de Datos (Z-score)

En este ejercicio, actuaremos como científicos de datos que necesitan preparar un conjunto de métricas antes de ingresarlas a un modelo de Machine Learning. La **estandarización** es un proceso que transforma los datos para que tengan una media de $0$ y una desviación estándar de $1$.

### Consigna
Debes desarrollar una serie de funciones en Python
<br> que transformen la siguiente lista de datos:
`datos = [5, 7, 10, 12, 15, 18, 20]`

### Fórmulas a utilizar

1. **Media ($\mu$):**
   $$\mu = \frac{\sum x_i}{n}$$

2. **Desviación Estándar Poblacional ($\sigma$):**
   $$\sigma = \sqrt{\frac{\sum (x_i - \mu)^2}{n}}$$

3. **Fórmula de Estandarización ($z$):**
   $$z = \frac{x - \mu}{\sigma}$$

---

### Requerimientos técnicos
* **Modularidad:** Crea tres funciones: `calcular_media(lista)`, `calcular_desviacion(lista)` y `estandarizar(lista)`.
* **Eficiencia:** Para el cálculo de la suma de los cuadrados de la diferencia en la desviación estándar, utiliza una **expresión generadora** (pista: `sum(... for x in lista)`).
* **Restricción:** No utilices librerías externas como `numpy` o `pandas`. El objetivo es practicar lógica pura de Python.

---

### Resultado esperado
Al procesar la lista `[5, 7, 10, 12, 15, 18, 20]` y redondear los resultados a **dos decimales**, deberías obtener exactamente estos valores:

* **-1.44**
* **-1.05**
* **-0.47**
* **-0.08**
* **0.50**
* **1.08**
* **1.47**

````python
# @title Mostrar la solución
def calcular_media(lista):
    """Calcula el promedio aritmético."""
    return sum(lista) / len(lista)


def calcular_desviacion(lista):
    """Calcula la desviación estándar poblacional."""
    # Llamamos a la función calcular_media
    mu = calcular_media(lista)

    # Suma de los cuadrados de las diferencias
    varianza = sum((x - mu) ** 2 for x in lista) / len(lista)

    return varianza ** 0.5 # elevamos a 0.5 (eso es aplicar raiz)


def estandarizar_datos(lista):
    """Transforma la lista a una distribución con media 0 y std 1."""
    mu = calcular_media(lista)
    sigma = calcular_desviacion(lista)

    # Aplicamos la fórmula (x - mu) / sigma a cada elemento
    lista_z = [(x - mu) / sigma for x in lista]
    return lista_z



# Datos de prueba
datos = [5, 7, 10, 12, 15, 18, 20]

# Ejecución
datos_z = estandarizar_datos(datos)

print("Datos originales:", datos)
print("Datos estandarizados (Z-scores):")
for val in datos_z:
    # el 5 es la cantidad de caracteres que deja para imprimir
    # el 2 es la cantidad de decimales para redondear
    print(f"{val:5.2f}")
````
