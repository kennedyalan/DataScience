## Selección de Datos: `loc` vs `iloc`

| Método | Selección por | Tipo de acceso | Sintaxis básica |
|--------|---------------|----------------|-----------------|
| **loc** | Etiquetas | Inclusivo | `df.loc[fila_inicio:fila_fin, col_inicio:col_fin]` |
| **iloc** | Índices numéricos | Exclusivo en fin | `df.iloc[fila_inicio:fila_fin+1, col_inicio:col_fin+1]` |

**Ejercicio práctico:** Seleccionar las primeras 5 filas y las columnas 'Age', 'Fare', 'Sex' usando ambos métodos.

```python
# Demostración loc
subset_loc = df.loc[:4, ['Age', 'Fare', 'Sex']]

# Demostración iloc
indices_columnas = [df.columns.get_loc(c) for c in ['Age', 'Fare', 'Sex']]
subset_iloc = df.iloc[:5, indices_columnas]

print("Resultado con loc:")
display(subset_loc)

print("\nResultado con iloc:")
display(subset_iloc)
```

## Normalización

$$
z = \frac{x - \mu}{\sigma}
$$

Donde:

- $x$ es el valor original  
- $\mu$ es la media de la variable  
- $\sigma$ es la desviación estándar  

La variable estandarizada $z$ tiene media $0$ y desviación $1$.



```python
from sklearn.preprocessing import StandardScaler

df_norm = df.copy()
cols_to_scale = ['Age', 'Fare', 'SibSp', 'Parch']

scaler = StandardScaler()

df_norm[cols_to_scale] = scaler.fit_transform(df_norm[cols_to_scale])

df_norm.head()
```

Escalar
$$
x' = \frac{x - \min(x)}{\max(x) - \min(x)}
$$

Donde:

- $x$ es el valor original  
- $\min(x)$ es el valor mínimo de la variable  
- $\max(x)$ es el valor máximo de la variable  

El valor normalizado $x'$ queda en el rango $[0, 1]$.


```python
from sklearn.preprocessing import MinMaxScaler

df_esca = df.copy()
cols_to_scale = ['Age', 'Fare', 'SibSp', 'Parch']

scaler = MinMaxScaler()

df_esca[cols_to_scale] = scaler.fit_transform(df_esca[cols_to_scale])

df_esca.head()
```

```python
# Reseteamos el indice sin guardar el viejo
df_esca = df_esca.reset_index(drop=True)
df_esca.info()
```

## Train Test Split

```python
import random

size = len(df_esca)
train_size = int(size * 0.8)

df_shuffled = df_esca.sample(frac=1, random_state=42).reset_index(drop=True)

train_df = df_shuffled.iloc[:train_size]
test_df  = df_shuffled.iloc[train_size:]

print(f"Tamaño del conjunto de entrenamiento: {len(train_df)}")
print(f"Tamaño del conjunto de prueba: {len(test_df)}")
```
