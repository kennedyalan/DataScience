## Constructores y Destructores

### Método __init__
El método `__init__` es el constructor, ejecutado automáticamente al crear un objeto. Inicializa los atributos de instancia.

### Método __del__
El método `__del__` es el destructor, invocado cuando un objeto es destruido, útil para liberar recursos.

## Encapsulamiento
El encapsulamiento protege los datos al restringir el acceso directo a los atributos, lo que permite controlar cómo se acceden y modifican desde fuera de la clase.

### Atributos Públicos y Privados
Por convención, los atributos que comienzan con un guion bajo (por ejemplo, _saldo) se consideran privados y no deben modificarse directamente desde fuera de la clase. Esto ayuda a mantener la integridad de los datos internos.

### Métodos Getter y Setter con @property y @setter
En Python, en lugar de utilizar métodos get_ y set_ tradicionales, se recomienda usar el decorador @property. Este permite acceder y modificar atributos privados como si fueran públicos, pero con validaciones internas.



``` python
# Ejemplo de un método get_ para acceder a la propiedad _nombre (si existe)
def get_nombre(self):
    return self._nombre
```



````python
class CuentaBancaria:
    def __init__(self, saldo_inicial):
        self._saldo = saldo_inicial

    @property
    def saldo(self):  # Getter
        return self._saldo

    @saldo.setter
    def saldo(self, nuevo_saldo):  # Setter con validación
        if nuevo_saldo >= 0:
            self._saldo = nuevo_saldo
        else:
            print("El saldo no puede ser negativo.")


# Uso
cuenta = CuentaBancaria(1000)
print(cuenta.saldo)    # Accede al saldo (getter)
cuenta.saldo = 1500    # Modifica el saldo (setter)

print(cuenta.saldo)    # Accede al saldo (getter)
cuenta.saldo = -500    # Muestra mensaje de error
````

En Python, nada es realmente privado como en otros lenguajes como Java o C++.


El guion bajo (_atributo) es una convención, no una restricción.



**¿Qué significa eso?**


Cuando escribimos:

```self._saldo = saldo_inicial```

**Estamos diciendo:**

*“Esto es privado por convención, no lo toques directamente a menos que sepas lo que estás haciendo”.*

**Pero Python no impide que hagamos:**
```python
cuenta = CuentaBancaria(1000)
print(cuenta._saldo)  #Acceso directo (no recomendado, pero posible)
```


**Y entonces para qué sirve?**


*   Para indicar al programador que ese atributo no debería ser tocado directamente.
*   Para usarlo internamente junto con @property y encapsular lógica (como validaciones).


**¿Y si queremos que sea más difícil de acceder?**


Podemos usar doble guion bajo, lo que activa un mecanismo llamado *name mangling*:

````python
class Persona:
    def __init__(self, nombre):
        self.__nombre = nombre
````

¿Podemos ahora acceder a `persona.__nombre`?

````python
persona = Persona("Ana")
print(persona.__nombre)
````

pero todavía podríamos acceder con:

````python
# Preguntar por los atributos y métodos
dir(persona)
````

````python
# y acceder
print(persona._Persona__nombre)
````

### Ejercicio Práctico
Crea una clase Libro que gestione el acceso a su atributo protegido `_titulo` mediante el uso de propiedades `@property`.

 El `"setter"` debe validar que el nuevo título no sea una cadena vacía antes de asignarlo.

#### Solución (desactivada)

````python
class Libro:
    def __init__(self, titulo):
        self._titulo = titulo

    @property
    def titulo(self):
        return self._titulo

    @titulo.setter # debe existir la property 'titulo'
    def titulo(self, nuevo_titulo):
      if len(nuevo_titulo) > 0:
          self._titulo = nuevo_titulo
      else:
        print("Título inválido")

# Uso
libro = Libro("El Quijote")
print(libro.titulo)  # Salida: El Quijote
print()

libro.titulo = ""
print()

print(libro.titulo)
````

## El polimorfismo
El polimorfismo es uno de los principios fundamentales de la programación orientada a objetos (OOP). Se refiere a la capacidad de diferentes clases de responder a la misma interfaz (método o función) de manera diferente. Es decir, el polimorfismo permite que una misma operación se realice de diferentes formas dependiendo del objeto que la invoque.


**Ejemplo en Python:**


Vamos a crear un ejemplo sencillo para ilustrar el polimorfismo. Supongamos que tenemos una clase base llamada Animal, y varias clases derivadas como Perro y Gato que sobrescriben un método común hacer_sonido(). Cada tipo de animal hace un sonido diferente, pero todos comparten la misma interfaz (el método hacer_sonido()).

**Python usa duck typing:**

*“Si camina como pato y hace cuac como pato… es pato.”*

````python
class Animal:
    def hacer_sonido(self):
        pass # Este método debe ser sobrescrito por una subclase

class Perro(Animal):
    def hacer_sonido(self):
        return "Guau"

class Gato(Animal):
    def hacer_sonido(self):
        return "Miau"

# Función que acepta cualquier objeto de tipo Animal y llama al método hacer_sonido()
def mostrar_sonido(animal):
    print(animal.hacer_sonido())

# Instanciamos los objetos
perro = Perro()
gato = Gato()

# Llamamos a la función mostrar_sonido para ambos
mostrar_sonido(perro)  # Guau
mostrar_sonido(gato)   # Miau
````

## Clases abstractas

Una clase abstracta es una clase que define una interfaz o comportamiento común, pero deja algunas partes sin implementar, obligando a las subclases a completar esas partes.

Se utiliza cuando queremos definir una estructura base general, pero sabemos que los detalles concretos los va a implementar cada subclase.

### Ejemplo:



``` python
from abc import ABC, abstractmethod

class Animal(ABC):  # Clase abstracta

    @abstractmethod
    def hacer_sonido(self):
        # No se implementa, solo se declara.
        # Las subclases deben implementar este método
        pass  

class Perro(Animal):
    def hacer_sonido(self):
        return "Guau"

class Gato(Animal):
    def hacer_sonido(self):
        return "Miau"

```



Ahora Python no deja instanciar subclases que no implementen el método.

### Ejercicio Práctico
Crea una clase abstracta `Empleado` con un método abstracto `calcular_salario`, y dos clases derivadas `EmpleadoTiempoCompleto` y `EmpleadoPorHoras`.

#### Solución (desactivada)

````python
from abc import ABC, abstractmethod

class Empleado(ABC):
    @abstractmethod
    def calcular_salario(self):
        pass


class EmpleadoTiempoCompleto(Empleado):
    def __init__(self, salario_mensual):
        self.salario_mensual = salario_mensual

    def calcular_salario(self):
        return self.salario_mensual


class EmpleadoPorHoras(Empleado):
    def __init__(self, horas_trabajadas, tarifa_hora):
        self.horas_trabajadas = horas_trabajadas
        self.tarifa_hora = tarifa_hora

    def calcular_salario(self):
        return self.horas_trabajadas * self.tarifa_hora


# Uso
empleados = [EmpleadoTiempoCompleto(3000), EmpleadoPorHoras(160, 20)]
for empleado in empleados:
    print(empleado.calcular_salario())  # Salida: 3000, 3200
````
